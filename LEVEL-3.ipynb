```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv('Dataset.csv')
df1.head()
```

```python
df1['Has Table booking'].replace({'Yes':1,'No':0},inplace=True)
df1['Has Online delivery'].replace({'Yes':1,'No':0},inplace=True)
df1['Is delivering now'].replace({'Yes':1,'No':0},inplace=True)
df1['Switch to order menu'].replace({'Yes':1,'No':0},inplace=True)


df1['Has Table booking'] = pd.to_numeric(df1['Has Table booking'])
df1['Has Online delivery'] = pd.to_numeric(df1['Has Online delivery'])
df1['Is delivering now'] = pd.to_numeric(df1['Is delivering now'])
df1['Switch to order menu'] = pd.to_numeric(df1['Switch to order menu'])
```

```python
df1.head()
```

Let's drop those columns which will not have siginificant effect or is not related  in deciding the Aggregate rating

```python
df2=df1.drop(['Restaurant ID','Cuisines','Rating color','Rating text','Currency','Restaurant Name','Country Code',	'City',	'Address',	'Locality'	,'Locality Verbose'],axis='columns')
df2.head()
```

Now let's split the dataset into training and testing with appropriate ratio.

```python
X=df2.drop(['Aggregate rating'],axis='columns')
X.head()
```

```python
y=df2['Aggregate rating']
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.2)
```

FIRST OF WE WILL USE MULTIPLE VARIABLE LINEAR REGRESSION AND CALCULATE THE MODEL SCORE

```python
from sklearn.linear_model import LinearRegression
```

```python
model1=LinearRegression()
```

```python
model1.fit(X_train ,y_train)
```

```python
model1.predict(X_test)
```

```python
model1.score(X_test , y_test)
```

```python
model1.predict([[140,18.57,1500,1,1,1,0,4,500]])
```

The score for multiple linear regression is nearly 0.29
